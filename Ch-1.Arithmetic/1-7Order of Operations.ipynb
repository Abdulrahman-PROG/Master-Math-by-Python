## Order of Operations

```python
4 + 3 * 2 - 1
```

```python
4+3  *  2-1
```

```python
(4 + 3) * 2 - 1
```

```python
4 + 3 * (2 - 1)
```

```python
4 + 3 * 5 / 6 + 3
```

```python
4 + 3 * (5 / 6 + 3)
```

```python
(4 + 3) * 5 / 6 + 3
```

```python
(4 + 3) * 5 / (6 + 3)
```
